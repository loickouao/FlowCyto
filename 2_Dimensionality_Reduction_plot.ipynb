```python
import pandas as pd
import numpy as np
from pandas.api.types import CategoricalDtype
from plotnine import *
from plotnine.data import mpg
%matplotlib inline

df_subsampleEmb = pd.read_csv('Resultats/2_Dimensionality_Reduction/2_result1_Dimension_Reduction.csv')  

df_subsampleEmb2 = pd.read_csv('Resultats/2_Dimensionality_Reduction/2_result2_Dimension_Reduction.csv')  

print("#We can also color this plot according to the expression values of a certain marker of interest:")

p1_CD19 = (ggplot(df_subsampleEmb)               # defining what data to use
+ aes(x='tsne1', y='tsne2', color='CD19') # defining what variable to use
+ geom_point()                         # defining the type of plot to use
+ theme_minimal()                      # defining a theme (optional)
+ labs(title='tsne 1000 cells data of channels of_interest, Expression CD3', x='Tsne1', y='Tsne2')
)
ggsave(plot=p1_CD19, filename='Resultats/2_Dimensionality_Reduction/Figure_2b - tsne 1000 cells of channels of_interest_CD19.png', dpi=100)
 

p1_CD3 = (ggplot(df_subsampleEmb)              
+ aes(x='tsne1', y='tsne2', color='CD3')
+ geom_point()                        
+ theme_minimal()                    
+ labs(title='tsne 1000 cells data of channels of_interest, Expression CD3', x='Tsne1', y='Tsne2')
)
ggsave(plot=p1_CD3, filename='Resultats/2_Dimensionality_Reduction/Figure_2c - tsne 1000 cells of channels of_interest_CD3.png', dpi=100)



print("#Run tSNE on the same subsample of cells, but this time using only the CD3, CD11c and CD19 channels.")

p2_CD19 = (ggplot(df_subsampleEmb2)              
+ aes(x='tsne1', y='tsne2', color='CD19')
+ geom_point()                        
+ theme_minimal()                    
+ labs(title='tsne 1000 cells data of 3 channels CD3, CD11c and CD19', x='Tsne1', y='Tsne2')
)

ggsave(plot=p2_CD19, filename='Resultats/2_Dimensionality_Reduction/Figure_3b - tsne 1000 cells of channels of_interest_CD19.png', dpi=100)


p2_CD3 = (ggplot(df_subsampleEmb2)              
+ aes(x='tsne1', y='tsne2', color='CD3')
+ geom_point()                        
+ theme_minimal()                    
+ labs(title='tsne 1000 cells data of 3 channels CD3, CD11c and CD19', x='Tsne1', y='Tsne2')
)

ggsave(plot=p2_CD3, filename='Resultats/2_Dimensionality_Reduction/Figure_3c - tsne 1000 cells of channels of_interest_CD3.png', dpi=100)

p2_CD3

```
